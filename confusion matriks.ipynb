```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

**Sample Data**

```python
file_path = 'Data Mahasiswa.csv'
data_mahasiswa = pd.read_csv(file_path)
data_mahasiswa
```

**Preprocessing**

```python
data_mahasiswa['JENIS KELAMIN'] = data_mahasiswa['JENIS KELAMIN'].map({'LAKI - LAKI': 0, 'PEREMPUAN': 1})
data_mahasiswa['STATUS MAHASISWA'] = data_mahasiswa['STATUS MAHASISWA'].map({'MAHASISWA': 0, 'BEKERJA': 1})
data_mahasiswa['STATUS PERNIKAHAN'] = data_mahasiswa['STATUS PERNIKAHAN'].map({'BELUM': 0, 'MENIKAH': 1})
data_mahasiswa['STATUS KELULUSAN'] = data_mahasiswa['STATUS KELULUSAN'].map({'TEPAT': 1, 'TERLAMBAT': 0})
data_mahasiswa
```

```python
X = data_mahasiswa[['JENIS KELAMIN', 'STATUS MAHASISWA', 'STATUS PERNIKAHAN', 'IPK SEMESTER 1-6']]
y = data_mahasiswa['STATUS KELULUSAN']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Confusion Matrix:")
print(conf_matrix)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
