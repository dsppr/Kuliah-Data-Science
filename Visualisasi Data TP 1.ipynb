# Visualisasi Data Tugas Personal 1 - Mohamad Rizki Septiyanto

**Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Data Frame**

```python
df = pd.read_csv('datasettp1.csv')
df
```

**Variabel yang dipakai**

```python
negara = df['Negara']
jumlah_penduduk = df['Jumlah Penduduk']
```

**Membuat Line Chart**

```python
plt.figure(figsize=(10, 6))
plt.plot(negara, jumlah_penduduk, marker='o', linestyle='-', color='b')
plt.title('Jumlah Penduduk Negara-Negara di Asia Tenggara')
plt.xlabel('Negara')
plt.ylabel('Jumlah Penduduk (juta)')
plt.grid(True)
plt.show()
```
