**Anda diberikan koin yang bias dengan peluang munculnya angka adalah 0.55, i.e.  X∼Bernoulli(0.55).**

a.	Simulasikan 2000 kali lemparan koin dengan bernoulli.rvs(). Simpanlah dalam variabel x

b.	Hitunglah nilai rata-rata kemunculan 1 dan 0 setelah 1 lemparan, 2 lemparan, ..., hingga 2000 lemparah
 
c.	Gambarkan perubahan nilai x¯ tersebut dalam diagram garis (petunjuk: akan ada dua garis). Apa yang dapat Anda amati?


**Simulasi 2000 Kali Lemparan Koin**

```python
import numpy as np
from scipy.stats import bernoulli

p = 0.55
x = bernoulli.rvs(p, size=2000)
x_labels = np.where(x == 1, 'Angka', 'Kepala')

print(x_labels[:2000])
```

**Menghitung Nilai Rata-rata Kemunculan "Angka" dan "Kepala"**

```python
cumulative_mean = np.cumsum(x) / np.arange(1, len(x) + 1)
mean_kepala = (np.arange(1, len(x) + 1) - np.cumsum(x)) / np.arange(1, len(x) + 1)

# Menyimpan hasil dalam DataFrame untuk kemudahan plotting
import pandas as pd
results = pd.DataFrame({
    'Lemparan': np.arange(1, len(x) + 1),
    'Rata-rata Angka': cumulative_mean,
    'Rata-rata Kepala': mean_kepala
})

print(results)
```

**Menggambarkan Perubahan Nilai Rata-rata**

```python
import matplotlib.pyplot as plt

# Membuat plot
plt.figure(figsize=(12, 6))
plt.plot(results['Lemparan'], results['Rata-rata Angka'], label='Rata-rata Kemunculan Angka', color='b')
plt.plot(results['Lemparan'], results['Rata-rata Kepala'], label='Rata-rata Kemunculan Kepala', color='r')

# Menambahkan judul dan label
plt.title('Perubahan Nilai Rata-rata Kemunculan Angka dan Kepala')
plt.xlabel('Jumlah Lemparan')
plt.ylabel('Nilai Rata-rata')
plt.legend()
plt.grid(True)

# Menampilkan plot
plt.show()
```
