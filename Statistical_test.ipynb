```python
#Data
import pandas as pd
df = pd.read_csv('dataset_penjualan_sayur.csv')
df
```

```python
#Memisahkan Data
tomat = df[df['Sayuran'] == 'Tomat']['Penjualan (kg)']
bayam = df[df['Sayuran'] == 'Bayam']['Penjualan (kg)']

print('Tomat:\n', tomat.describe())
print('Bayam:\n', bayam.describe())
```

```python
# T-test
from scipy import stats

t_stat, p_value = stats.ttest_ind(tomat, bayam)

print('T-statistic:', t_stat)
print('P-value:', p_value)

#Hasil
alpha = 0.05
if p_value < alpha:
    print("Terdapat perbedaan signifikan antara penjualan Tomat dan Bayam.")
else:
    print("Tidak terdapat perbedaan signifikan antara penjualan Tomat dan Bayam.")
```

```python
#Visualisasi
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(tomat, bins=5, color='tomato', alpha=0.7)
plt.title('Distribusi Penjualan Tomat')
plt.xlabel('Penjualan (kg)')
plt.ylabel('Frekuensi')

plt.subplot(1, 2, 2)
plt.hist(bayam, bins=5, color='green', alpha=0.7)
plt.title('Distribusi Penjualan Bayam')
plt.xlabel('Penjualan (kg)')
plt.ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```
